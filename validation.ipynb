```python
from pymongo import MongoClient

# Connect to the MongoDB server (localhost by default)
mg_client1 = MongoClient('mongodb://localhost:27017/', username='root', password='rootpwd')

##provide feedback to the user
if mg_client1:
    print("connected to mongo database successully.")

else:
    print("failed to connect to mongo database.")
```

```python
# get db

db = mg_client1["socialmedia"]


# set validation for post collection
validation_rule = {
    "$jsonSchema": {
        "bsonType": "object",
        "required": ["PostID", "UserID", "Content", "Timestamp"],
        "properties": {
            "PostID": {"bsonType": "string"},
            "UserID": {"bsonType": "string"},
            "PostType": {"bsonType": "string"},
            "Content": {"bsonType": "string"},
            "Timestamp": {"bsonType": "date"},
            "Comments": { # embedded comments
                "bsonType": "array",
                "items": {
                    "bsonType": "object",
                    "required": ["CommentID", "UserID", "CommentText"],
                    "properties": {
                        "CommentID": {"bsonType": "string"},
                        "UserID": {"bsonType": "string"},
                        "CommentText": {"bsonType": "string"},
                        "Timestamp": {"bsonType": "string"}
                    }
                }
            },
            "Hashtags": { # embedded hashtags
                "bsonType": "array",
                "items": {
                    "bsonType": "object",
                    "required": ["HashtagID"],
                    "properties": {
                        "HashtagID": {"bsonType": "string"}
                    }
                }
            }, # likes/share
            "LikeCount": {"bsonType": "int"},
            "ShareCount": {"bsonType": "int"}
        }
    }
}

db.command({
    "collMod": "posts",
    "validator": validation_rule,
    "validationLevel": "moderate"
})


```
